```python
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.svm import SVC
import time

image_dir = Path("/content/fish_images")
image_dir2= Path("/content/mask_images")
# Get filepaths and labels
filepaths = sorted(list(image_dir.glob(r'**/*.png')))
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepaths))

maskpaths=sorted(list(image_dir2.glob(r'**/*.png')))
filepaths = pd.Series(filepaths, name='Filepath').astype(str)
labels = pd.Series(labels, name='Label')
mask=pd.Series(maskpaths, name='maskedimage').astype(str)

# Concatenate filepaths and labels
image_df = pd.concat([filepaths, labels,mask], axis=1)

print(image_df['maskedimage'][27369])
```

```python
features=[]
labels=[]
red=[]
blue=[]
green=[]
edges=[]
Laplasian=[]
for i in range(1,len(image_df),1):
  file=image_df['maskedimage'][i]
  img1=cv2.imread(file)
  file1=image_df['Filepath'][i]
  img2=cv2.imread(file1)
    
  img3=cv2.subtract(img1,img2)
  plt.imshow(img3)
  img3=resize(img3,(80,80))

  img3=img3.flatten()
  features.append(img3)
  #print(features.shape)
    
for label in image_df['Label']:
    labels.append(label)
    
    
```

```python

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.25)
n_components = 15
pca = PCA(n_components=n_components).fit(X_train)
#eigenfaces = pca.components_.reshape((n_components, h, w))
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
#print(principalDf)
plt.plot(pca.explained_variance_ratio_)
plt.show()
```

```python
model=SVC(C=1,kernel='poly',gamma='auto')
```

```python
model.fit(X_train,y_train)
prediction=model.predict(X_test)
accuracy=model.score(X_test,y_test)
#Labels_final=['Fishes1','Fishes2','Fishes3','Fishes4']
print(y_test)
print(accuracy)



```
